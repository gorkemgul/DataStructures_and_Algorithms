# BigO Notation

* BigO Notation -> 0()

```python
# define a function for BigOn
def bigOn(n):
    for i in range(0, n):
        print(i)
```

```python
# let's test our BigOn function
bigOn_func = bigOn(5)
bigOn_func
```

```python
# define a function for BigOn2
def bigOn2(n):
    for i in range(0, n):
        for j in range(0, n):
            print(i, j)
```

```python
# Now, let's test our BigOn2 function
bigOn2_func = bigOn2(5)
```

```python
# define a function for BigOn3
def bigOn3(n):
    for i in range(0, n):
        for j in range(0, n):
            for k in range(0, n):
                print(i, j, k)
```

```python
# let's try our BigOn3 function
bigOn3_func = bigOn3(5)
```

```python
# Let's create a logn function
import math
def logn(n):
    while n > 1:
        n = math.floor(n / 2)
        print(n)
```

```python
# let's try our logn function
logn_func = logn(32)
logn_func
```

```python
# define a nlogn function
def nlogn(n):
    lim = n
    while n > 1:
        n = math.floor(n / 2)
        for i in range(1, lim):
            print(i)
```

```python
# let's try our nlogn function
nlogn_func = nlogn(16)
nlogn_func
```

```python
# define a nfactorial function
def nfactorial(n):
    if n == 0:
        print('1')
    else:
        for i in range(0, n):
            print(i)
            nfactorial(n-1)
```

```python
# let's try our nfactorial function
nfactorial_func = nfactorial(5)
```

```python
# None of those functions that we've written above are about calculating factorial! 
# Now, let's just create a function to calculate actual factorial

counter = 0
def actualFactorial(n):
    global counter
    counter += 1
    print(counter)
    if (n == 1 or n == 0):
        return 1
    else:
        return (n * actualFactorial(n - 1))
```

```python
actual_factorial_func = actualFactorial(5)
actual_factorial_func
```
